```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter


#defining all the basic trigonometric functions

def func1(x):
    return np.sin(x)*3     # sin function   #here amplitude is multiplies by 3 for better view in graph

def func2(x):
    return np.cos(x)*3    #cosine function   #amplitude multiplied by 3

def func3(x):
    return (np.sin(x)/np.cos(x)) # tan function

def func4(x):
    return np.cos(x)/np.sin(x)   #cot function

def func5 (x):
    return 1/(np.sin(x))      # cosec function

def func6(x):
    return 1/(np.cos(x))      #sec function



metadata=dict(title='Trigonometric fucntions animated',artist='Mario')
writer=PillowWriter(fps=15, metadata=metadata)




#creating a layout of subplots

fig,ax=plt.subplots(3,2,figsize=(10,9))      # three rows and two columns
fig.suptitle("Trigonometric Functions",fontweight='bold') #top most title 


xlist=[]
ylist1=[]
ylist2=[]
ylist3=[]
ylist4=[]
ylist5=[]
ylist6=[]

#plotting each plot individually

l1,=ax[0,0].plot([],[],color='green')  # empty plot at 0,0 whre l1 will be plotted
ax[0,0].set_title("sine")              #setting the title for plot at 0,0
ax[0,0].set_xlim([-5,5])       # setting the limit for x axis
ax[0,0].set_ylim([-5,5])       # setting the limit for y axis

l2,=ax[0,1].plot([],[],color='blue')
ax[0,1].set_title('cosine')
ax[0,1].set_xlim([-5,5])
ax[0,1].set_ylim([-5,5])

l3,=ax[1,0].plot([],[],color='red')
ax[1,0].set_title('tan')
ax[1,0].set_xlim([-5,5])
ax[1,0].set_ylim([-5,5])


l4,=ax[1,1].plot([],[],color='yellow')
ax[1,1].set_title('cot')
ax[1,1].set_xlim([-5,5])
ax[1,1].set_ylim([-5,5])


l5,=ax[2,0].plot([],[],color='brown')
ax[2,0].set_title('cosec')
ax[2,0].set_xlim([-5,5])
ax[2,0].set_ylim([-5,5])


l6,=ax[2,1].plot([],[],color='magenta')
ax[2,1].set_title('sec')
ax[2,1].set_xlim([-5,5])
ax[2,1].set_ylim([-5,5])




with writer.saving(fig,"trigofunc_animation.gif",100):    # saving  the results in a gif file with name "triginofunc_animation" 
                                                          # 100 states the dpi(resolution)
    for xval in np.linspace(-5,5,100):   # iterating through a set of x values
        xlist.append(xval)
        ylist1.append(func1(xval))       # updating empty lists declared above by adding values to them.
        ylist2.append(func2(xval))
        ylist3.append(func3(xval))
        ylist4.append(func4(xval))
        ylist5.append(func5(xval))
        ylist6.append(func6(xval))
        
        l1.set_data(xlist,ylist1)        #plotting the lines 
        l2.set_data(xlist,ylist2)
        l3.set_data(xlist,ylist3)
        l4.set_data(xlist,ylist4)
        l5.set_data(xlist,ylist5)
        l6.set_data(xlist,ylist6)
        
        writer.grab_frame()   #Grab the image information from the figure and save as a movie frame.
        




plt.show()
```
